```python
import random as rnd
import math
import heapq as hq
def Read(filename):
    G = []
    i = 0
    file = open(filename)
    for line in file:
        G.append([float(x) for x in line.split('\t')])
        
        
    return G

CP = Read('dataset.txt')
CP2 = Read('dataset2.txt')
CP3 = Read('dataset3.txt')
CP4 = Read('dataset4.txt')

def distancia(point1, point2):
    return ((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2) ** 0.5
```

```python
def Grafito(CPoblados):
    n = len(CPoblados)
    G = [[] for _ in range(n)]
    N = []
    for _ in range(n):
        N.append(_)
    j = -1
    for i in CPoblados:
        auxiliar = CPoblados[:]
        auxiliar.remove(i)
        CantDestNodo = rnd.randint(6,7)
        j +=1
        for _ in range(CantDestNodo):
            dest = rnd.choice(auxiliar)
            auxiliar.remove(dest)
            dist = distancia(i,dest)
            conexion = (dist, int(CPoblados.index(dest)))
            G[j].append(conexion)
    return G


def dijkstra(G, s):
    n = len(G)
    visited = [False]*n
    weights = [math.inf]*n
    path = [None]*n
    queue = []
    weights[s] = 0
    c = 0
    hq.heappush(queue, (0, s))
    while len(queue) > 0:
        g, u = hq.heappop(queue)
        visited[u] = True
        for w, v in G[u]:
            if not visited[v]:
                f = g + w
                if f < weights[v]:
                    weights[v] = f
                    path[v] = u
                    c += w
                    hq.heappush(queue, (f, v))
    return path, weights, c
```

```python
print(dijkstra(Grafito(CP),0))
print("\n\n\n")
print(dijkstra(Grafito(CP2),0))
print("\n\n\n")
print(dijkstra(Grafito(CP3),0))
```
