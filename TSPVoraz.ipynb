```python
import math
def Read(filename):
    G = []
    i = 0
    file = open(filename)
    for line in file:
        G.append([float(x) for x in line.split('\t')])
        
    return G
```

```python
print(Read('dataset.txt'))
```

```python
def distancia(nodo1, nodo2):
    return ((nodo1[0] - nodo2[0])**2 + (nodo1[1] - nodo2[1])**2) ** 0.5

def tsp(G, s=None):
    if s is None:
        s = G[0]
    por_visitar = G
    camino = [s]
    por_visitar.remove(s)
    while por_visitar:
        proximo = min(por_visitar, key=lambda x: distancia(camino[-1], x))
        camino.append(proximo)
        por_visitar.remove(proximo)
    camino.append(s)
    return camino

G = Read('dataset.txt')
print(tsp(G)) 
```
