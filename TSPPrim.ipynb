```python
import random as rnd
import math
import heapq as hq
def Read(filename):
    G = []
    i = 0
    file = open(filename)
    for line in file:
        G.append([float(x) for x in line.split('\t')])
                
    return G

CP = Read('dataset.txt')
CP2 = Read('dataset2.txt')
CP3 = Read('dataset3.txt')
CP4 = Read('dataset4.txt')

def distancia(point1, point2):
    return ((point1[0] - point2[0])**2 + (point1[1] - point2[1])**2) ** 0.5
```

```python
def Grafito(CPoblados):
    n = len(CPoblados)
    G = [[] for _ in range(n)]
    N = []
    for _ in range(n):
        N.append(_)
    j = -1
    for i in CPoblados:
        auxiliar = CPoblados[:]
        auxiliar.remove(i)
        CantDestNodo = rnd.randint(4,7)
        j +=1
        for _ in range(CantDestNodo):
            dest = rnd.choice(auxiliar)
            auxiliar.remove(dest)
            dist = distancia(i,dest)
            conexion = (dist, int(CPoblados.index(dest)))
            G[j].append(conexion)
    return G

def prim(G):
    n = len(G)
    dist = [math.inf]*n
    path = [-1]*n
    visited = [False]*n
    q = []
    hq.heappush(q, (0, 0))
    dist[0] = 0
    while len(q) > 0:
        _, u = hq.heappop(q)
        if not visited[u]:
            visited[u] = True
            for w, v in G[u]:
                if not visited[v] and w < dist[v]:
                    dist[v] = w
                    path[v] = u
                    hq.heappush(q, (w, v))

    return path, dist
```

```python
print(prim(Grafito(CP)))
print("\n\n\n")
print(prim(Grafito(CP2)))
print("\n\n\n")
print(prim(Grafito(CP3)))
```
